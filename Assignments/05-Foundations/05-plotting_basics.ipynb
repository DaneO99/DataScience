## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

```python
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

plt.plot(x1, y1, label="sin(x)")
plt.plot(x2, y2, label="cos(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x) vs cos(9)")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

#set 1
x1 = np.random.rand(30)
y1 = np.random.rand(30)
sizes1 = 300 * np.random.rand(30)
colors1 = np.random.rand(30)

#set 2
x2 = np.random.rand(20)
y2 = np.random.rand(20)
sizes2 = 300 * np.random.rand(20)
colors2 = np.random.rand(20)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.scatter(x1, y1, s=sizes1, c=colors1, alpha=0.6, cmap='viridis')
plt.scatter(x2, y2, s=sizes2, c=colors2, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example with My 2 sets")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
categories = ["Golf", "Movies","Video Games","Work"]
values = [2,4,6,8]
plt.bar(categories,values)
plt.title("Time spent for each activity")
plt.xlabel("Activity")
plt.ylabel("Time (hours)")
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```
